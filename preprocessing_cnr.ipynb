```python
import pandas as pd
import os
import sys
import numpy as np
import re
from datetime import datetime
```

```python
raw_df = pd.read_csv("data/raw_data/X_train_v2.csv")
raw_df_test = pd.read_csv("data/raw_data/X_test_v2.csv")
raw_df_y = pd.read_csv("data/raw_data/Y_train_sl9m6Jh.csv")
```

```python
raw_df.head(60)
```

```python
wip_df = raw_df.query("WF == 'WF1'").copy()
wip_df['dt'] = pd.to_datetime(wip_df['Time'],format='%d/%m/%Y %H:%M') - pd.DateOffset(hours=1)
wip_df['date'] = wip_df['dt'].dt.date
wip_df['hour'] = wip_df['dt'].dt.hour
final_df = wip_df[['ID','dt','date','hour']].merge(raw_df_y,how='left',on='ID',validate='one_to_one')

wip_df_test = raw_df_test.query("WF == 'WF1'").copy()
wip_df_test['dt'] = pd.to_datetime(wip_df_test['Time'],format='%d/%m/%Y %H:%M') - pd.DateOffset(hours=1)
wip_df_test['date'] = wip_df_test['dt'].dt.date
wip_df_test['hour'] = wip_df_test['dt'].dt.hour
final_df_test = wip_df_test[['ID','dt','date','hour']].copy()
```

```python
def combine_forecasts(wip_df,final_df):
    for nwp_ix in range(1,5):
        nwp_df = wip_df.filter(like='NWP' + str(nwp_ix)).copy()

        print("######")
        print("NWP: " + str(nwp_ix))

        for fcst_item in ['U','V','T','CLCT']:
            #nwp_df.filter(regex= "_" + fcst_item + "$")
            print("------")
            print("Item: " + str(fcst_item))
            fcst_item_df = nwp_df.filter(regex= "_" + fcst_item + "$")
            print("Ncol: " + str(len(list(fcst_item_df.columns))))
            if len(list(fcst_item_df.columns)) == 0:
                continue

            the_col_name = 'NWP'+ str(nwp_ix) +'_' + fcst_item 
            final_df[the_col_name + '_train'] = np.NaN
            final_df[the_col_name + '_test'] = np.NaN

            ## 00h_D-1
            allowed_for_test = True
            for col in fcst_item_df.columns:
                print("Column: " + col)
                final_df[the_col_name + '_train'] = nwp_df[col].fillna(final_df[the_col_name + '_train'])
                if allowed_for_test:
                    final_df[the_col_name + '_test'] = final_df[the_col_name + '_train']
               # print(re.search("00h_D-1_",col)
                if re.search("00h_D-1_",col) is not None:
                    print("End of test")
                    allowed_for_test = False

        print("######")
    return final_df
```

```python
train_df = combine_forecasts(wip_df,final_df)
test_df = combine_forecasts(wip_df_test,final_df_test)
```

```python
train_df.to_excel("data/cnr_train_data.xlsx",index=False)
test_df.to_excel("data/cnr_test_data.xlsx",index=False)
```

```python
train_raw_data = train_df.copy()
train_wip_data = train_raw_data.set_index('dt').copy()
nwp_col = [col for col in train_wip_data if col.startswith('NWP')]
train_wip_data[nwp_col] = train_wip_data[nwp_col].interpolate(method='time')
train_cols = ['ID','date','hour','Production'] + [col for col in train_wip_data if col.endswith('_train')] 
train_cols_proper = [col.replace("_train","") for col in train_cols]
train_cols_proper = map(lambda x:x.lower(),train_cols_proper)
train_final_df = train_wip_data[train_cols].copy()
train_final_df.columns = train_cols_proper

train_final_df.to_excel("data/preprocessed_train_data.xlsx",index=True)
```
